```python
import math 
import numpy as np
class Hash:
    def __init__(self, name, tableLength):
        self.flag = 0
        self.name = name
        self.employeesNum = 0
        self.tableLength = tableLength
        self.A = (math.sqrt(5) - 1)/2
        self.table = np.array([None for i in range(self.tableLength)])
        for i in range(self.tableLength):
            self.table[i] = []
        print("The hash-table %s has been created!" % self.name)
    
    def __hashing(self, key):
        # В качестве хеш-функции я выбрал метод умножения, в котором
        # h(key) = M * ((A * key) mod 1), где A лежит в интервале [0;1],
        # M - размер таблицы. В качестве A я взял число (sqrt(5) - 1)/2,
        # т.е. число, обратное к золотому сечению (взято из книги Кнута).
        result = 0
        key = self.__strToInt(key)
        result = self.tableLength*(math.modf(key*self.A)[0])
        return int(result)
    
    def __strToInt(self, key):
        Sum = 0
        key = list(key)
        for i in range(len(key)):
            if(key[i] != ' '):
                Sum += ord(key[i])
        return Sum
    
    def search(self, entity):
        if self.check(entity):
            key = entity[0]
            value = entity[1]
            hashValue = self.__hashing(key)
            if (self.table.all()):
                if not self.flag:
                    print("The element doesn't exist")
                self.flag = 0
                return 0
            elif (not (self.table[hashValue])):
                if not self.flag:
                    print("The element doesn't exist")
                self.flag = 0
                return 0
            else:
                return self.__listSearch(self.table[hashValue], entity)
        
    def __listSearch(self, list, entity):
        if not list:
            if not self.flag:
                print("The element doesn't exist")
            self.flag = 0
            return 0
        elif not list.count(entity):
            if not self.flag:
                print("The element doesn't exist")
            self.flag = 0
            return 0
        else:
            if not self.flag:
                print("Employee's position is", list[list.index(entity)][1])
            return list[list.index(entity)][1]
        
    def add(self, entity):
        self.flag = 1
        if not self.search(entity):
            key = entity[0]
            value = entity[1]
            hashValue = self.__hashing(key)
            self.table[hashValue].append(entity)
            print("The element is successfuly added")
        else:
            print("You are trying to add, but the element already exists")
        self.flag = 0
    
    def remove(self, entity):
        self.flag = 1
        if self.search(entity):
            key = entity[0]
            value = entity[1]
            hashValue = self.__hashing(key)
            self.table[hashValue].remove(entity)
            print("The element is successfully removed")
        else:
            print("You are trying to remove, but the element doesn't exist!")
        self.flag = 0
            
    def check(self, entity):
        try:
            if(type(entity)==tuple and type(entity[0]) == str and type(entity[1]) == str and len(entity) == 2):
                return 1
            else:
                raise TypeError
        except TypeError as tp:
            print("TypeError! {0}".format(tp))
            return 0           
```
